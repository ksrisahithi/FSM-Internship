# Model definition

```python
class NaiveBayes:
    def train(self,X,y):
        samples,features = X.shape
        self._classes = np.unique(y).astype(np.uint32)
        classes = len(self._classes)
        
        self.mean = np.zeros((classes,features), dtype=np.float64)
        self.var = np.zeros((classes,features), dtype=np.float64)
        self.prior = np.zeros(classes,dtype=np.float64)
        
        for c in self._classes:
#             c = int(c)
            X_c = X[c==y]
            self.mean[c,:] = X_c.mean(axis=0)
            self.var[c,:] = X_c.var(axis=0)
            self.prior[c] = X_c.shape[0] / float(samples)
        
    def test(self,X):
        y_pred = [self._predict(x) for x in X]
        return y_pred
    
    def _predict(self,x):
        posteriors = []
        
        for i,c in enumerate(self._classes):
            prior = np.log(self.prior[i])
            class_conditional = np.sum(np.log(self._pd(i,x)))
            posterior = prior + class_conditional
            posteriors.append(posterior)
            
        return self._classes[np.argmax(posteriors)]
            
    def _pd(self, class_ind, x):
        mean = self.mean[class_ind]
        var = self.var[class_ind]
#         print(x)
#         print('xshape: {}, meanshape: {}, varShape: {}, xtype: {}, meantype: {}, vartype: {}'.format(x.shape, mean.shape, var.shape, x.dtype, mean.dtype, var.dtype))
        numerator = np.exp(-(x-mean)**2 / (2*var))
        denominator = np.sqrt(2*np.pi*var)
        return numerator / denominator
```

# Training 

```python
X_train = inputs.to_numpy(dtype=np.float64)
y_train = target.to_numpy(dtype=np.float64)
nb = NaiveBayes()
nb.train(X_train,y_train)
```
